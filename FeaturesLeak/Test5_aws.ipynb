```python
import FeaturesLeak as fl
%matplotlib inline
```

```python
grid_cv = fl.StratifiedKFold(n_splits=10,
                          shuffle=True,
                          random_state=0)

eval_cv = fl.StratifiedKFold(n_splits=10,
                          shuffle=True,
                          random_state=1)
```

```python
myClass = fl.ChooseSubsection(grid_cv, eval_cv, 
                            k_tr            = 3,
                            k_bag           = 3,
                            test_size       = 0.1,
                            collect_n       = 10,
                            size_dim        = 264,
                            num_iteration   = 400,
                            learning_rate   = 0.4,
                            print_log       = True,
                            max_vec         = 15)
```

```python
myClass.init_data()
```
